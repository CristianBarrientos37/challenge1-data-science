### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# prompt: Análisis de facturación

# Para cada tienda, vamos a calcular:
# - Ingreso total
# - Número de facturas
# - Promedio de ingreso por factura

# Función para analizar una tienda
def analizar_tienda(df, nombre_tienda):
    ingreso_total = df['Precio'].sum() # Changed 'Importe' to 'Precio'
    numero_facturas = df.shape[0] # Using number of rows as a proxy for number of invoices
    promedio_ingreso_por_factura = ingreso_total / numero_facturas if numero_facturas > 0 else 0

    print(f"--- Análisis de {nombre_tienda} ---")
    print(f"Ingreso Total: {ingreso_total:,.2f}")
    print(f"Número de Facturas: {numero_facturas}")
    print(f"Promedio de Ingreso por Factura: {promedio_ingreso_por_factura:,.2f}")
    print("-" * (len(f"--- Análisis de {nombre_tienda} ---")))

# Analizar cada tienda
analizar_tienda(tienda, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```

# 2. Ventas por categoría

```python
# prompt: Ventas por categoría

# Unir los dataframes para analizar todas las tiendas juntas
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Calcular las ventas por categoría
ventas_por_categoria = todas_tiendas.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("\nVentas por Categoría (Total de todas las tiendas):")
ventas_por_categoria
```

```python
# prompt: ahora separa esas ventas por categoria por cada tienda

# Función para analizar ventas por categoría por tienda
def analizar_ventas_por_categoria_por_tienda(df, nombre_tienda):
    print(f"\n--- Ventas por Categoría en {nombre_tienda} ---")
    ventas_categoria = df.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
    print(ventas_categoria)
    print("-" * (len(f"--- Ventas por Categoría en {nombre_tienda} ---")))

# Analizar ventas por categoría para cada tienda
analizar_ventas_por_categoria_por_tienda(tienda, "Tienda 1")
analizar_ventas_por_categoria_por_tienda(tienda2, "Tienda 2")
analizar_ventas_por_categoria_por_tienda(tienda3, "Tienda 3")
analizar_ventas_por_categoria_por_tienda(tienda4, "Tienda 4")
```

# 4. Productos más y menos vendidos

```python
# prompt: Productos más y menos vendidos por tienda

# Función para encontrar los productos más y menos vendidos por tienda
def productos_mas_menos_vendidos_por_tienda(df, nombre_tienda):
  print(f"\n--- Productos más y menos vendidos en {nombre_tienda} ---")
  # Agrupar por 'Nombre del Producto' y contar las ocurrencias
  conteo_productos = df['Producto'].value_counts()

  if not conteo_productos.empty:
    # Producto más vendido (el que aparece más veces)
    producto_mas_vendido = conteo_productos.index[0]
    cantidad_mas_vendido = conteo_productos.iloc[0]
    print(f"Producto más vendido: '{producto_mas_vendido}' ({cantidad_mas_vendido} unidades)")

    # Producto menos vendido (el que aparece menos veces)
    producto_menos_vendido = conteo_productos.index[-1]
    cantidad_menos_vendido = conteo_productos.iloc[-1]
    print(f"Producto menos vendido: '{producto_menos_vendido}' ({cantidad_menos_vendido} unidades)")
  else:
    print("No hay datos de productos para analizar.")

  print("-" * (len(f"--- Productos más y menos vendidos en {nombre_tienda} ---")))

# Analizar productos más y menos vendidos para cada tienda
productos_mas_menos_vendidos_por_tienda(tienda, "Tienda 1")
productos_mas_menos_vendidos_por_tienda(tienda2, "Tienda 2")
productos_mas_menos_vendidos_por_tienda(tienda3, "Tienda 3")
productos_mas_menos_vendidos_por_tienda(tienda4, "Tienda 4")

# Para el total de todas las tiendas
print("\n--- Productos más y menos vendidos (Total de todas las tiendas) ---")
conteo_productos_total = todas_tiendas['Producto'].value_counts()

if not conteo_productos_total.empty:
  producto_mas_vendido_total = conteo_productos_total.index[0]
  cantidad_mas_vendido_total = conteo_productos_total.iloc[0]
  print(f"Producto más vendido: '{producto_mas_vendido_total}' ({cantidad_mas_vendido_total} unidades)")

  producto_menos_vendido_total = conteo_productos_total.index[-1]
  cantidad_menos_vendido_total = conteo_productos_total.iloc[-1]
  print(f"Producto menos vendido: '{producto_menos_vendido_total}' ({cantidad_menos_vendido_total} unidades)")
else:
  print("No hay datos de productos para analizar.")
print("-" * (len("--- Productos más y menos vendidos (Total de todas las tiendas) ---")))
```

# 5. Envío promedio por tienda

```python
# prompt: Envío promedio por tienda

# Función para calcular el envío promedio por tienda
def calcular_envio_promedio(df, nombre_tienda):
  if 'Costo de envío' in df.columns:
    envio_promedio = df['Costo de envío'].mean()
    print(f"Envío promedio en {nombre_tienda}: {envio_promedio:.2f}")
  else:
    print(f"La columna 'Costo de envío' no se encuentra en el dataframe de {nombre_tienda}.")

# Calcular envío promedio para cada tienda
print("\n--- Envío promedio por tienda ---")
calcular_envio_promedio(tienda, "Tienda 1")
calcular_envio_promedio(tienda2, "Tienda 2")
calcular_envio_promedio(tienda3, "Tienda 3")
calcular_envio_promedio(tienda4, "Tienda 4")
print("-" * (len("--- Envío promedio por tienda ---")))
```
